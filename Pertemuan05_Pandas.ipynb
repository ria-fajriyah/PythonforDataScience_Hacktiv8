### Objek Seri
Dalam panda, data satu dimensi direpresentasikan sebagai objek seri. Objek seri diinisialisasi menggunakan pd.Series() konstruktor

```python
series2 = pd.Series([[10,20],[30,40, 'Series'],[50,60], {'name':'ria', 'org':'community'}])
```

```python
print(series2)
```

```python
vars = pd.Series()
print(vars)
```

```python
datanp = np.array(['x','y','z','a','b','c','d','e','f'])
dataser = pd.Series(datanp)
print(dataser)
print(dataser.index)
```

```python
datanp2 = np.array(['x','y','z','a','b','c'])
dataser2 = pd.Series(datanp2, index=[20,21,22,23,24,25])
print(dataser2)
print(dataser2.index)
```

```python
data_dic = {'A':0.7, 'B':0, 'C':10.5, 'D':1}
pds = pd.Series(data_dic)
print(pds)
print(pds.index)
```

### DataFrames
DataFrame adalah representasi data dua dimensi dalam baris dan kolom, yang dapat diinisalisasi dalam panda menggunakan DataFrame()konstruktor

```python
data_dict = {
        'name' : ['Joko','Jono','Joni','Jhon'],
        'age' : [20,50,30,40],
        'address' : ['JKT','BDG','SBY','SDA'],
        'qualif' : ['M.Sc','Prof','Ir','Ph.D']
}
print(data_dict)
```

```python
df = pd.DataFrame(data_dict)
print(df)
```

```python
print(df[['name','qualif']])
```

```python
print(df['qualif'] == 'Ph.D')
```

```python
kondisiphd = df['qualif'] == 'Ph.D'
```

```python
print(df.loc[kondisiphd])
```

```python
dfangka = pd.DataFrame({'x':[2,3],'y':[10,20]})
print(dfangka)
```

```python
dfapl = dfangka.apply(lambda z: z*3)
print(dfapl)
```

```python
def square(a):
    return a*a
```

```python
print(dfangka)
```

```python
dfsquare = dfangka.apply(square)
print(dfsquare)
```
