## Python for Loop

### a. Looping in List

```python
languages = ['Golang', 'Python', 'Java', 'C#', 'Rust', 'Vlang']
print(languages)

for lang in languages:
    print("bahasa pemrograman:")
    print(lang)
```

### b. Looping in Dictionary

```python
mahasiswa = {"NIM":"205150201111035", "Nama":"Nurul Fajriyah Al Maulidiyah", "Nilai":90}
for mhs in mahasiswa:
    print(mhs)
    print(mahasiswa[mhs])
```

```python
for k, v in mahasiswa.items():
    print(k, ":", v)
```

### c. Range Function

```python
values = range(5)
print(values)
for x in values:
    print(x)
```

```python
for x in (0,1,2,3,4,5):
    print(x)
```

```python
angka = range(5)
for i in angka:
    if i % 2 == 0:
        print(i, ": Genap")
    else:
        print(i, ": Ganjil")
```

```python
digits = [0,1,2,3,4,5]
for i in digits:
    print(i)
else:
    print("no item left")
```

## Aftering for Loop Behavior
### Break

```python
for i in range(5):
    print("Loop in range")
    if i == 3:
        break
    
    print("Cetak angka ke",i)
```

```python
i = 1
for varlist  in ['foo','bar','baz','qmux']:
    print(varlist)
    if 'z' in varlist:
        break
    
    print(i)
    i +=1
    
print(i) 
```

### Continue

```python
for varlist  in ['foo','bar','baz','qmux']:
    print(varlist)
    if 'r' in varlist:
        print("r ditemukan")
        continue
    
    print("Kondisi dibawah continue")
    
print("Selesai")
```

```python
i = 1 
n =5 
while i <= n:
    print(i)
    i +=1 
```

```python
counter = 1
while counter < 4:
    print("didalam while")
    counter += 1
else:
    print("inside while else")    
```

```python
def SwitchExamples(args):
    switch = {
        0 : "Angka Nol",
        1 : "Angka Satu",
        2 : "Angka 2"
    }
    return switch.get(args, "Nothing" )
```

```python
hasil = SwitchExamples(0)
print(hasil)
```

```python
hasil = 0 
temp = input("Ketik temparatur yang ingin di konversi eg. 40F, 18C, 0C: ")

print(type(temp))
print(temp)
degree = int(temp[:-1])
print("degree : ",degree)
print(f"format degree : {degree} ")

convertion = temp[-1]
print("Convertion : ",convertion)

if convertion == 'C':
    print("Celcius")
    hasil = int(round(9 * degree) / 5 + 32)
elif convertion == 'F':
    print("F")
    hasil = int(round(degree - 32) * 5 / 9 )
else: 
    print("Yang anda input salah")
    
print("Tempartur adalah : ", hasil, "derajat")
```
